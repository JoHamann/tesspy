```python
from tessellation import Tessellation
```

# Create Tessllation Object

```python
ffm= Tessellation('Frankfurt am Main')
```

```python
ffm.get_polygon().plot().set_axis_off();
```

# Squares and Hexagons

```python
ffm_hex=ffm.hexagons(7)
ffm_hex.plot(lw=1, edgecolor='w').set_axis_off();
```

```python
ffm_sqr=ffm.squares(14)
ffm_sqr.plot(lw=1, edgecolor='w').set_axis_off();
```

# Voronoi

```python
ffm_voronoi= ffm.voronoi(cluster_algo='k-means',
                        poi_categories=['amenity', 'shop'],
                        timeout=60,
                        n_polygons=500,
                        verbose=True)
```

```python
ffm_voronoi.plot(lw=1, edgecolor='w').set_axis_off();
print(len(ffm_voronoi), 'polygons')
```
