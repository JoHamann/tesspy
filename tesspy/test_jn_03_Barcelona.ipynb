```python
from tessellation import Tessellation
```

# Create Tessllation Object

```python
barcelona= Tessellation('barcelona')
```

```python
barcelona_poly=barcelona.get_polygon()
```

```python
barcelona_poly
```

```python
barcelona_poly.plot().set_axis_off();
```

# Squares and Hexagons

```python
barcelona_hex=barcelona.hexagons(8)
barcelona_hex.plot(lw=1, edgecolor='w').set_axis_off();
```

```python
barcelona_sqr=barcelona.squares(16)
barcelona_sqr.plot(lw=1, edgecolor='w').set_axis_off();
```

# Voronoi

```python
barcelona_voronoi= barcelona.voronoi(cluster_algo='k-means',
                        poi_categories=['public_transport'],
                        timeout=60,
                        n_polygons=500,
                        verbose=True)
```

```python
barcelona_voronoi.plot(lw=1, edgecolor='w').set_axis_off();
print(len(barcelona_voronoi), 'polygons')
```

```python
barcelona_voronoi= barcelona.voronoi(cluster_algo='k-means',
                        poi_categories=['amenity', 'shop'],
                        timeout=60,
                        n_polygons=100,
                        verbose=True)
```

```python
barcelona_voronoi.plot(lw=1, edgecolor='w').set_axis_off();
print(len(barcelona_voronoi), 'polygons')
```
